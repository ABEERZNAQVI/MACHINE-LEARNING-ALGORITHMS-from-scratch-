# Randomized K-Means for MNIST dataset

This code implements a simplified k-Means algorithm for image classification using the MNIST dataset. The process involves randomly initializing cluster centers for each digit class and assigning test images to the class with the nearest cluster center.<br>The code uses Pandas for data manipulation and NumPy for numerical operations.

```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
dg=pd.read_csv("mnist_train.csv")
df=dg.copy()
dt= pd.read_csv("mnist_test.csv")
y_test_label  = dt['label'].to_numpy()
```

```python
train1_data = np.zeros((60000,785))
for i in range(len(df)):
    train1_data[i] = df.iloc[i][:]   
```

```python
test1_data = np.array(np.zeros((10000,785)))
for i in range(len(dt)):
    test1_data[i] = dt.iloc[i][:]
```

```python
mean={}
for i in range (10):
    kp = []
    indexs = list(df.loc[df["label"]==i].index)
    for j in range(5):
        random.shuffle(indexs)
        clust = train1_data[random.choices(indexs,k=round(0.2*len(indexs))),1:]
        kp.append(clust.mean(axis=0))
    mean[i]=kp
```

```python
pred=[]
for i in test1_data[:,1:]:
    d={}
    for j in mean:
        for k in mean[j]:
            d[np.sqrt(np.sum((i-k)**2))]=j
    u = [d[t] for t in sorted(d)[:5]]
    pred.append(max(u,key=u.count))
pred
```

```python
cor,incor=0,0
for i in range(len(pred)):
    if pred[i]==y_test_label[i]:
        cor+=1
    else:
        incor+=1
accuracy=np.multiply(np.divide(cor,len(test1_data)),100)
accuracy
```
